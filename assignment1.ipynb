1.Why are dictionary keys immutable in Python? Explain with an example.
Answer:
Python dictionaries use a hash-based structure to store keyâ€“value pairs.
When a key is inserted, its hash value is calculated once and used for fast access.
If a key were mutable and changed later, its hash value would also change, and Python would no longer be able to locate it.
Therefore, dictionary keys must be immutable to ensure consistency and efficiency.

2. Write a program to count how many times each element occurs in a list.
Answer:
The program iterates through the list and maintains a dictionary to store how many times each element appears.

```python
# List containing elements
data_list = [1, 2, 2, 3, 4, 4, 4, 5]

occurrence = {}

for element in data_list:
    occurrence[element] = occurrence.get(element, 0) + 1

print("Element frequencies:")
print(occurrence)
```

3. Explain the difference between dict[key] and dict.get(key) with an example.
Answer:
dict[key] returns the value of a key but raises a KeyError if the key does not exist.
dict.get(key) returns the value if the key is present; otherwise, it returns None or a default value without causing an error.
Thus, get() is safer when the key may not be available.

```python
info = {
    "name": "Aakanksha",
    "roll_no": 101
}

# Unsafe method
# print(info["age"])   # KeyError

# Safe method
print("Age using get():", info.get("age"))

# Using default value
print("Age with default:", info.get("age", "Not Found"))
```

4. Write a program to merge two dictionaries by keeping the larger value for common keys.
Answer:
If a key exists in both dictionaries, the program compares their values and stores the larger value in the final dictionary.
Keys present in only one dictionary are added directly.

```python
first_dict = {"a": 10, "b": 20, "c": 30}
second_dict = {"b": 25, "c": 15, "d": 40}

final_dict = {}

for k, v in first_dict.items():
    final_dict[k] = v

for k, v in second_dict.items():
    if k in final_dict:
        final_dict[k] = v if v > final_dict[k] else final_dict[k]
    else:
        final_dict[k] = v

print("Final merged dictionary:")
print(final_dict)
```

5. Create a dictionary using comprehension where keys are odd numbers and values are their cubes.
